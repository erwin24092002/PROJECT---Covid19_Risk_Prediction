```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pickle
import time
from sklearn.linear_model import LogisticRegression
```

**EXPERIMENT**

solver='liblinear', penalty='l1'

```python
for i in range(1, 4):
  print('-'*30, f'[BỘ {i}]','-'*30)

  data_train = pd.read_csv(f'./lan{i}/train.csv')
  data_test = pd.read_csv(f'./lan{i}/test.csv')

  X_train, y_train = data_train.iloc[:,:-1], data_train.iloc[:,-1]
  X_test, y_test = data_test.iloc[:,:-1], data_test.iloc[:,-1]

  model = LogisticRegression(solver='liblinear' ,penalty='l1', random_state = 2508)
  start = time.time()
  model.fit(X_train, y_train)
  print('time = ', time.time()-start)

  y_pred = model.predict(X_test)

  print('precision score = ', precision_score(y_test, y_pred))
  print('recall score = ', recall_score(y_test, y_pred))
  print('f1 score = ', f1_score(y_test, y_pred))
  print('accuracy = ', accuracy_score(y_test, y_pred))
```
